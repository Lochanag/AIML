```python
import pandas as pd
#read csv file into Dataframe
df=pd.read_csv('/content/drive/MyDrive/drive/fifa-world-cup (1).csv')
df['YEAR'] = df['YEAR'].map(lambda x:x*3.10)
print(df)
```

```python
filtered_df = df[df['MATCHES'] > 20]
print(filtered_df)
```

```python
from functools import reduce
if not filtered_df['MATCHES'].empty:
   total = reduce(lambda x,y:x+y,filtered_df['MATCHES'])
   print(total)
else:
  print("The filtered DataFrame is empty,cannot calculate total MATCHES.")
```

```python
import pandas as pd
df=pd.read_csv('/content/drive/MyDrive/drive/fifa-world-cup (1).csv')
import matplotlib.pyplot as plt
df.groupby('MATCHES')['GOALS'].sum().plot(kind='bar')
plt.title("BAR PLOT")
plt.xlabel('MATCHES')
plt.ylabel('GOALS')
plt.show()
```
